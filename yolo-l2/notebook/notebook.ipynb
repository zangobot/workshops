# YOLO Adversarial Challenge Solver
**Goal:** Generate an image that is visually similar to `source_image.jpg` ($L_2$ distance < 50) but fails to be detected as the original object by YOLOv11. Then make it specifically a gibbon.

```python
import numpy as np
from PIL import Image
from ultralytics import YOLO
import matplotlib.pyplot as plt

# 1. Load the Model
model = YOLO("yolo11n-cls.pt")

# 2. Load the Reference Image
original_image = Image.open("nika.png").convert("RGB")
original_array = np.array(original_image, dtype=np.float32)

print(f"Image Loaded. Shape: {original_array.shape}")
plt.imshow(original_image)
plt.axis('off')
plt.show()
```

```python
def top_class(results):
    r = results[0]
    top_idx = int(r.probs.top1)
    conf = float(r.probs.top1conf)
    class_name = r.names[top_idx]
    return (class_name, conf)
results = model(original_image)
top_class(results)
```

```python
adversarial_image = original_image.copy()
```

```python
adv_array = np.array(adversarial_image, dtype=np.float32)
diff = adv_array - original_array
l2_dist = np.linalg.norm(diff)

print(f"Current L2 Distance: {l2_dist:.2f}")

if l2_dist > 50:
    print("⚠️ WARNING: L2 Distance is too high!")
else:
    print("✅ L2 Distance is within limits.")
```

```python
import requests
import io, os

# Load URLs from environment variables
URL_UNTARGETED = os.environ.get("VERIFICATION_SERVER_UNTARGETED")
URL_TARGETED = os.environ.get("VERIFICATION_SERVER_TARGETED")

def submit_solution(image_obj, mode="untargeted"):
    """
    Submits the image to the verification server.
    mode: "untargeted" (default) or "targeted"
    """
    # Select the correct URL
    if mode == "targeted":
        api_url = URL_TARGETED
        print(f"🚀 Submitting to TARGETED Server: {api_url}")
    else:
        api_url = URL_UNTARGETED
        print(f"🚀 Submitting to UNTARGETED Server: {api_url}")

    if not api_url:
        print(f"⚠️ Error: {mode.upper()} server URL is not set in environment variables.")
        return

    buf = io.BytesIO()
    image_obj.save(buf, format="PNG")
    buf.seek(0)
    
    # Send POST request
    try:
        response = requests.post(api_url, files={'image': buf}, timeout=5)
        data = response.json()
        
        if data.get("success"):
            print("🎉 VICTORY!")
            print(f"Flag: {data['flag']}")
            print(f"Final L2 Distance: {data.get('l2_distance', 'N/A'):.2f}")
            if 'predicted_class' in data:
                print(f"Server saw Class: {data['predicted_class']} (Conf: {data.get('confidence', 0):.2f})")
        else:
            print("❌ Attempt Failed")
            print(f"Message: {data['message']}")
            if 'l2_distance' in data:
                print(f"L2 Distance: {data['l2_distance']:.2f}")
            if 'predicted_class' in data:
                print(f"Server saw Class: {data['predicted_class']} (Conf: {data.get('confidence', 0):.2f})")
                
    except requests.exceptions.ConnectionError:
        print(f"⚠️ Error: Could not connect to {api_url}.")
    except Exception as e:
        print(f"⚠️ Error: {e}")

# Usage Examples:
submit_solution(adversarial_image, mode="untargeted")
submit_solution(adversarial_image, mode="targeted")
```
